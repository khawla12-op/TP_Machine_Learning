## Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="darkgrid")
```

## Plotting functions :

### Relpot :

```python
tips = sns.load_dataset("tips")
tips
```

```python
sns.relplot(data=tips, y="tip", x="total_bill")
plt.show()
```

```python
sns.relplot(data=tips, y="tip", x="total_bill", hue="smoker")
plt.show()
```

```python
sns.relplot(data=tips, y="tip", x="total_bill", hue="smoker", style="smoker")
plt.show()
```

```python
sns.relplot(data=tips, y="tip", x="total_bill", hue="smoker", style="time")
plt.show()
```

```python
sns.relplot(data=tips, y="tip", x="total_bill", hue="size")
plt.show()
```

```python
sns.relplot(data=tips, y="tip", x="total_bill", hue="size", palette="ch:r=-.5,l=.75")
plt.show()
```

```python
sns.relplot(data=tips, y="tip", x="total_bill", size="size")
plt.show()
```

```python
sns.relplot(data=tips, y="tip", x="total_bill", size="size", sizes=(15, 200))
plt.show()
```

### Line plot

```python
flights=sns.load_dataset("flights")
flights
```

```python
may_flights=flights.query("month=='May'")
sns.lineplot(data=may_flights,x="year",y="passengers")
plt.show()
```

```python
flights_wide=flights.pivot("year","month","passengers")
flights_wide.head()
```

```python
sns.lineplot(data=flights_wide["May"])
plt.show()
```

### Plotting subsets of data with semantic mappings:

```python
fmri=sns.load_dataset("fmri")
fmri
```

```python
sns.relplot(x="timepoint", y="signal", kind="line", data=fmri)
plt.show()
```

```python
sns.relplot(x="timepoint", y="signal", hue="event", kind="line", data=fmri)
```

```python
sns.relplot(x="timepoint", y="signal", hue="region", style="event",kind="line", data=fmri)
plt.show()
```

## Plotting with Categorical Data:

```python
tips = sns.load_dataset("tips")
tips
```

```python
sns.catplot(x="day", y="total_bill", data=tips)
plt.show()
```

```python
sns.catplot(x="day", y="total_bill", jitter=False, data=tips)
plt.show()
```

```python
sns.catplot(x="smoker", y="tip", order=["No", "Yes"], data=tips)
plt.show()
```

```python
sns.catplot(x="total_bill", y="day", hue="time", kind="swarm", data=tips)
plt.show()
```

## BoxPlots:

```python
sns.catplot(x="day", y="total_bill", kind="box", data=tips)
plt.show()
```

```python
sns.catplot(x="day", y="total_bill", hue="smoker", kind="box", data=tips)
plt.show()
```

```python
tips["weekend"] = tips["day"].isin(["Sat", "Sun"])
sns.catplot(x="day", y="total_bill", hue="weekend",kind="box", dodge=False, data=tips)
plt.show()
```

## Annotated Heatmaps:

```python
flights_long = sns.load_dataset("flights")
flights_long
```

```python
flights = flights_long.pivot("month", "year", "passengers")
flights
```

```python
f, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(flights, annot=True, fmt="d", linewidths=.5, ax=ax)
```

## Multi-Plot Grid:

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
g = sns.FacetGrid(tips, col="time")
```

```python
g = sns.FacetGrid(tips, col="time")
g.map(sns.histplot, "tip")
```

```python
g = sns.FacetGrid(tips, col="sex", hue="smoker")
g.map(sns.scatterplot, "total_bill", "tip", alpha=.7)
g.add_legend()
```

```python
g = sns.FacetGrid(tips, row="smoker", col="time", margin_titles=True)
g.map(sns.regplot, "size", "total_bill", color=".3", fit_reg=False, x_jitter=.1)
```
